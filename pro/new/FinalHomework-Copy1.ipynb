## 算法代码（请在代码的关键环节做好注释）

```python
import copy

class TuoResult:
    def __init__(self, course_index, score):
        self.course_index = course_index
        self.score = score

class TuoList:
    def __init__(self):
        self.s = []
        self.cnt = 0

course_id = [[''] * 5 for _ in range(100)]
score = [0] * 100
G = [[0] * 100 for _ in range(100)]
term = 0
credit = 0
course_sum = 0

def prf_tidu(tidu, tidu_sum):
    for i in range(tidu_sum):
        print(f"Course: {course_id[tidu.s[i].course_index]}, Score: {tidu.s[i].score}")

def prf_course_fen(course_fen, term1):
    for i in range(term1):
        print(f"Term {i + 1} courses: ", end="")
        j = 0
        while course_fen[i][j] >= 0:
            print(f"{course_id[course_fen[i][j]]} ", end="")
            j += 1
        print()

def judge_course_id_index(string):
    for i in range(course_sum):
        print(course_id[i])
        if string == "".join(course_id[i]):
            return i
    return -1

def scaf(temp):
    global term, credit, course_sum
    term = int(input("Enter the number of terms: "))
    credit = int(input("Enter the maximum credits per term: "))
    course_sum = int(input("Enter the total number of courses: "))
    course_id = [[] for _ in range(course_sum)]  # 添加这一行
    print("Enter the name and score of each course:")
    for i in range(course_sum):
        name, s = input(f"Enter the name and score of course {i + 1}: ").split()
        score[i] = int(s)
        for j, c in enumerate(name):
            course_id[i].append(c)
    print("Enter the prerequisite relationships:")
    for i in range(course_sum):
        while True:
            s = input(f"Enter the prerequisite course for {course_id[i]} (enter 0 to finish): ")
            if s == '0':
                break
            index = judge_course_id_index(s)
            if index != -1:
                G[i][index] = 1
                temp[index] += 1
            else:
                print("Invalid course ID. Please try again.")

def add_e(temp):
    s1 = input("Enter the first course: ")
    s2 = input("Enter the prerequisite course: ")

    index1 = judge_course_id_index(s1)
    index2 = judge_course_id_index(s2)

    if index1 != -1 and index2 != -1:
        G[index1][index2] = 1
        temp[index2] += 1
        print("Prerequisite relationship added successfully.")
    else:
        print("Invalid course ID. Please try again.")

def crtl_z_e(temp):
    s1 = input("Enter the first course: ")
    s2 = input("Enter the prerequisite course: ")

    index1 = judge_course_id_index(s1)
    index2 = judge_course_id_index(s2)

    if index1 != -1 and index2 != -1:
        G[index1][index2] = 0
        temp[index2] -= 1
        print("Prerequisite relationship removed successfully.")
    else:
        print("Invalid course ID. Please try again.")

def temp_copy(temp, copy_tp):
    for i in range(course_sum):
        copy_tp[i] = temp[i]

def tuopu_sort(tidu, copy_tp, topo_sum):
    tidu.cnt = 0
    flag = False

    while True:
        for i in range(course_sum):
            if copy_tp[i] == 0:
                copy_tp[i] = -1
                tidu.s[tidu.cnt] = TuoResult(i, score[i])
                tidu.cnt += 1

                for j in range(course_sum):
                    if G[i][j] != 0:
                        copy_tp[j] -= 1

                flag = True

        if not flag:
            break

        flag = False

    if tidu.cnt != course_sum:
        print("There is a course loop.")
    else:
        print("The topological sorting result is:")
        prf_tidu(tidu, topo_sum)

def course_plan_xijie(tidu, course_fen, term1):
    max_score = 0
    for i in range(course_sum):
        if score[i] > max_score:
            max_score = score[i]

    for i in range(term1):
        max_sum = 0
        max_i = 0
        for j in range(tidu.cnt):
            if tidu.s[j].score > max_sum and course_fen[i][tidu.cnt] == -1 and score[j] != max_score:
                max_sum = tidu.s[j].score
                max_i = j

        course_fen[i][tidu.cnt] = tidu.s[max_i].course_index

def course_plan1(tidu, course_fen, term1):
    for i in range(term1):
        for j in range(tidu.cnt):
            if course_fen[i][j] == -1:
                course_fen[i][j] = tidu.s[j].course_index
                break

def course_plan2(tidu, course_fen, term1):
    k = 0
    for i in range(tidu.cnt):
        if course_fen[k][tidu.cnt] != -1:
            k += 1
        course_fen[k][i] = tidu.s[i].course_index

def prf_file(course_fen, term1):
    with open("course_schedule.txt", "w") as file:
        for i in range(term1):
            file.write(f"Term {i + 1} courses: ")
            j = 0
            while course_fen[i][j] >= 0:
                file.write(f"{course_id[course_fen[i][j]]} ")
                j += 1
            file.write("\n")

def dele_course(temp):
    s = input("Enter the name of the course you want to delete: ")
    index = judge_course_id_index(s)

    if index != -1:
        course_id[index] = [''] * 5
        score[index] = 0

        for i in range(course_sum):
            G[index][i] = 0
            G[i][index] = 0
            temp[i] -= 1

        print("Course deleted successfully.")
    else:
        print("Invalid course ID. Please try again.")

def dele_course_f(course_fen, tidu, term1):
    s = input("Enter the name of the course you want to delete: ")
    index = judge_course_id_index(s)

    if index != -1:
        for i in range(term1):
            for j in range(tidu.cnt):
                if course_fen[i][j] == index:
                    course_fen[i][j] = -1

        print("Course deleted successfully.")
    else:
        print("Invalid course ID. Please try again.")

def add_course(temp):
    global course_sum
    name, s = input("Enter the name and score of the new course: ").split()
    score[course_sum] = int(s)
    for i, c in enumerate(name):
        course_id[course_sum][i] = c

    for i in range(course_sum):
        G[i][course_sum] = 0
        G[course_sum][i] = 0

    print("Enter the prerequisite relationships:")
    for i in range(course_sum):
        while True:
            s = input(f"Enter the prerequisite course for {course_id[i]} (enter 0 to finish): ")
            if s == '0':
                break
            index = judge_course_id_index(s)
            if index != -1:
                G[i][index] = 1
                temp[index] += 1
            else:
                print("Invalid course ID. Please try again.")

    course_sum += 1
    print("Course added successfully.")

def menu():
    tidu = TuoList()
    course_fen = [[-1] * 100 for _ in range(10)]
    temp = [0] * 100
    copy_tp = [0] * 100
    topo_sum = 0
    term1 = 0
    menu_choice = 1

    while menu_choice:
        print("Course Management System:")
        print("1. Create Course Information")
        print("2. Add Prerequisite Relationships")
        print("3. Remove Prerequisite Relationships")
        print("4. Add Courses")
        print("5. Delete Courses")
        print("6. Topological Sorting")
        print("7. Generate Course Schedule")
        print("0. Exit")

        menu_choice = int(input("Enter your choice: "))

        if menu_choice == 1:
            scaf(temp)
            topo_sum = copy.copy(course_sum)
        elif menu_choice == 2:
            add_e(temp)
        elif menu_choice == 3:
            crtl_z_e(temp)
        elif menu_choice == 4:
            add_course(temp)
        elif menu_choice == 5:
            dele_course(temp)
        elif menu_choice == 6:
            temp_copy(temp, copy_tp)
            tuopu_sort(tidu, copy_tp, topo_sum)
        elif menu_choice == 7:
            print("1. Course Scheduling (with Credit Limits)")
            print("2. Course Scheduling (without Credit Limits)")
            print("3. Generate Course Schedule to File")
            print("4. Delete Courses from Schedule")
            print("0. Back")

            sub_choice = int(input("Enter your choice: "))

            if sub_choice == 1:
                term1 = int(input("Enter the number of terms: "))
                temp_copy(temp, copy_tp)
                tuopu_sort(tidu, copy_tp, topo_sum)
                course_plan_xijie(tidu, course_fen, term1)
                prf_course_fen(course_fen, term1)
            elif sub_choice == 2:
                term1 = int(input("Enter the number of terms: "))
                temp_copy(temp, copy_tp)
                tuopu_sort(tidu, copy_tp, topo_sum)
                course_plan1(tidu, course_fen, term1)
                prf_course_fen(course_fen, term1)
            elif sub_choice == 3:
                term1 = int(input("Enter the number of terms: "))
                temp_copy(temp, copy_tp)
                tuopu_sort(tidu, copy_tp, topo_sum)
                course_plan2(tidu, course_fen, term1)
                prf_file(course_fen, term1)
            elif sub_choice == 4:
                dele_course_f(course_fen, tidu, term1)
            elif sub_choice == 0:
                continue
            else:
                print("Invalid choice. Please try again.")
        elif menu_choice == 0:
            print("Exiting...")
            break
        else:
            print("Invalid choice. Please try again.")

menu()
```
