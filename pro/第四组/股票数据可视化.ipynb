```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import mplfinance as mpf
```

```python
# 读取数据
df = pd.read_excel('stock.xlsx')
# 设置字体
plt.rcParams['font.family'] = 'SimHei'
df
```

```python
# 1.
# 绘制K线图
fig, ax = plt.subplots()
ax.set_title('中国华融股票K线图')
ax.plot(df['交易日期'], df['最低价(元)'], 'g-', label='最低价')
ax.plot(df['交易日期'], df['收盘价(元)'], 'r-', label='收盘价')
ax.plot(df['交易日期'], df['最高价(元)'], 'b-', label='最高价')
ax.legend()
ax.grid(True)
fig.autofmt_xdate()
```

```python
import matplotlib.pyplot as plt
import mpl_finance as mpf
import pandas as pd

# 读取数据
df = pd.read_excel('stock.xlsx')
plt.rcParams['font.family'] = 'SimHei'

# 绘制 K 线图
fig,ax = plt.subplots(figsize=(12, 8))
mpf.candlestick2_ochl(ax, df['开盘价(元)'], df['收盘价(元)'], df['最高价(元)'], df['最低价(元)'],width=0.5,colorup='r',colordown='g',alpha=1.0)

# 设置图表标题、坐标轴标签等
ax.set_title('K Line Chart',fontsize=20)
ax.set_xlabel('交易日期',fontsize=14)
ax.set_ylabel('Price',fontsize=14)

plt.show()
```

```python
# 2.折线图
# 绘制开盘价和收盘价的线图
plt.plot(df['交易日期'], df['开盘价(元)'], label='开盘价')
plt.plot(df['交易日期'], df['收盘价(元)'], label='收盘价')

# 设置标题和坐标轴标签
plt.title('中国华融股票价格趋势')
plt.xlabel('日期')
plt.ylabel('价格')

# 添加图例
plt.legend()

# 显示图形
plt.show()
```

```python
# 3
# 设置画布大小
fig, ax = plt.subplots(figsize=(12, 6))

# 绘制成交量面积图
ax.fill_between(df['交易日期'], df['成交量(万股)'], alpha=0.3)
ax.set_xlabel('日期')
ax.set_ylabel('成交量（万股）')

# 修改y轴刻度
yticks = ax.get_yticks()
ax.set_yticks(yticks[::2])

plt.show()
```

```python
# 4
# 绘制涨跌幅图
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(df['交易日期'], df['涨跌幅'], color='red', alpha=0.8)
ax.axhline(y=0, color='black', linestyle='--', linewidth=0.5)
ax.set_xlabel('日期')
ax.set_ylabel('涨跌幅 %')
ax.set_title('涨跌幅')
ax.invert_yaxis()
plt.show()
```
