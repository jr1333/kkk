```python
import pandas as pd

# 读取 Excel 文件
df = pd.read_excel('merged.xlsx')
df.head()
```

```python
# 定义要检查的关键词列表
keywords = ['more', 'easier', 'better', 'clearer', 'smoother', 'enhance', 'build up', 'improve', 'optimize', 'upgrade', 'experience', 'update', 'UI', 'appropriateness', 'operability', 'protection']

# 创建一个空列，用于存储匹配结果
df['Result'] = 0

# 检查每行的 A 列是否包含关键词，并设置匹配行的 Result 列为 1
for index, row in df.iterrows():
    if any(keyword in str(row['Commit Message']) for keyword in keywords):
        df.at[index, 'Result'] = 1

# 将结果保存到新的 Excel 文件
df.to_excel('merged.xlsx', index=False)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

# 筛选出 'Result' 列为 1 的数据
filtered_df = df[df['Result'] == 1]

# 获取 'Commit Message' 列的文本数据
texts = filtered_df['Commit Message'].tolist()

# 创建 TfidfVectorizer 对象
vectorizer = TfidfVectorizer()

# 对文本数据进行 TF-IDF 向量化
tfidf_matrix = vectorizer.fit_transform(texts)

# 运行 K-means 聚类算法
kmeans = KMeans(n_clusters=2)
kmeans.fit(tfidf_matrix)

# 获取聚类的标签
labels = kmeans.labels_

# 获取聚类中心的关键词
feature_names = vectorizer.get_feature_names()
clusters_keywords = []
for i in range(kmeans.n_clusters):
    cluster_keywords = [feature_names[j] for j in kmeans.cluster_centers_[i].argsort()[:-6:-1]]
    clusters_keywords.append(cluster_keywords)

# 打印每个聚类的关键词集合
for i, keywords in enumerate(clusters_keywords):
    print(f"Cluster {i+1} Keywords: {', '.join(keywords)}")
```
